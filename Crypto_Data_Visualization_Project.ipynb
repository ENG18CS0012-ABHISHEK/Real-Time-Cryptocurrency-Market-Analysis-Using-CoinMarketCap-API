```python
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json
import os 
import pandas as pd 
# from time import time 
from time import sleep


pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

# Extracting Data Using Api From CoinMarketCap

```python
def api_runner():
    global df

    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    parameters = {
    'start':'1',
    'limit':'15',
    'convert':'USD'
    }
    headers = {
    'Accepts': 'application/json',
    'X-CMC_PRO_API_KEY': '47718b3d-4a53-4425-8ba1-0a6e68ce413c',
    }

    session = Session()
    session.headers.update(headers)

    try:
        response = session.get(url, params=parameters)
        data = json.loads(response.text)
        # print(data)
    except (ConnectionError, Timeout, TooManyRedirects) as e:
        print(e)

    df = pd.json_normalize(data['data'])
    df['time_stamp']= pd.to_datetime('now')
    df

    if not os.path.isfile(r'E:\Web_scrapping_Project\Crypto_website_api_project.csv'):
        df.to_csv(r'E:\Web_scrapping_Project\Crypto_website_api_project.csv', header='column_names')
    else :
        df.to_csv(r'E:\Web_scrapping_Project\Crypto_website_api_project.csv', mode='a', header=False)

 
```

```python
api_runner()
```

```python
for i in range(333):
    api_runner()
    print("Api completed successfully")
    sleep(60)
exit()
```
